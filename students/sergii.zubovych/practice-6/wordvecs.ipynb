```python
import glob
from langdetect import detect

def is_uk(filename):
    with open(filename) as f:
        text = f.read()
        if text.strip():
            return detect(text) == 'uk'
        return False

files = [f for f in glob.glob("1551/*") if is_uk(f)]

files[:10]
```

```python
import numpy as np
from nltk.probability import FreqDist

with open('uk_stopwords.txt') as f:
    STOP_WORDS = f.read().split()

def read_words(file):
    return list(filter(lambda w: w.strip() and w.isalpha(), re.split(r'[\,\;\.\"\!\?\s\:\-]', file.read())))

def file_2_vec(filename):
    with open(filename) as f:
        words = [w.lower().strip() for w in read_words(f)]
        fdist = FreqDist(words)
        vecs = []
        for w in words:
            if w in STOP_WORDS:
                continue
            try:
                vecs.append(wv[w]/fdist.freq(w))
            except KeyError as e:
                #print(w)
                pass
            
        return np.sum(vecs, axis=0)/len(vecs)
```

```python
vec = file_2_vec('1551/Відсутність-опалення-в-навчальних-закладах.txt')
```

```python
wv.most_similar(positive=[vec], topn=10)
```

```python
docvecs = []
for file in files:
    vec = file_2_vec(file)
    docvecs.append((file, vec, wv.most_similar(positive=[vec], topn=1)[0]))

docvecs
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=0).fit([vec[1] for vec in docvecs])
```

```python
print(kmeans.cluster_centers_)  
```

```python
[wv.most_similar(positive=[c], topn=1) for c in kmeans.cluster_centers_]
```

```python
topic_vecs = kmeans.predict([vec[1] for vec in docvecs])
```
